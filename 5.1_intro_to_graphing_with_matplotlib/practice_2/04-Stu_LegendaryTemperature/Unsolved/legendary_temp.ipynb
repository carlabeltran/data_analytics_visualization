```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define axes
avg_temp_f = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
avg_temp_c = [(temp - 32) * 0.56 for temp in avg_temp_f]
x_axis = np.arange(1, 13, 1)
```

```python
fahrenheight, = plt.plot(x_axis, avg_temp_f, marker="+", color="blue", label="Fahrenheit")
celsius, = plt.plot(x_axis, avg_temp_c, marker="s", color="red", label="Celsius" )
```

```python
plt.xlabel("Months")
plt.ylabel("Degress")
plt.legend(loc="best")
```
