```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define y-axis
avg_temp_f = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Define x-axis
x_axis = np.arange(1, 13, 1)

x_axis
```

```python
plt.plot(x_axis, avg_temp_f)
```

```python
# Convert temperature to degrees Celsius
avg_temp_c = [(temp - 32) * 0.56 for temp in avg_temp_f]

avg_temp_c
```

```python
# Plot the degrees Celsius points
plt.plot(x_axis, avg_temp_c)
```

```python
# Plot with both Celsius and Fahrenheit
plt.plot(x_axis, avg_temp_f)
plt.plot(x_axis, avg_temp_c)
plt.xlabel("Month")
plt.ylabel("Degrees")
plt.title("Monthly averages for temperature in New Jersey")
```
