```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define axes
danger_drop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
rail_gun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
x_axis = np.arange(0, 130, 10)
```

```python
plt.plot(x_axis, danger_drop, color="r", label="Danger Drop")
plt.plot(x_axis, rail_gun, color="g", label="Rail Gun")
```

```python
# Show grid lines
plt.grid()
```

```python
# x-axis label, y-axis label, and title
plt.xlabel("Coaster Runtime")
plt.ylabel("Speed (MPH)")
plt.title("Coaster Speed Over Time")
```

```python
# Add legend
plt.legend(loc="best")
```

```python
# Axis limits
plt.ylim(5, 95)
plt.xlim(0, 120)
```
