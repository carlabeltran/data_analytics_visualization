```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
danger_drop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]

rail_gun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]

x_axis = np.arange(0, 130, 10)

x_axis
```

```python
danger_drop_handle, = plt.plot(x_axis, danger_drop, color="red", label="Danger Drop")
rail_gun_handle = plt.plot(x_axis, rail_gun, color="green", label="Rail Gun")
```

```python
# grid lines
plt.grid()
```

```python
# axis labels
plt.xlabel("Coaster Runtime")
plt.ylabel("Speed (MPH)")
plt.title("Coaster Speed Over Time")
```

```python
# Add legend
plt.legend(loc="upper right")
```
