```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# y-axis - average temperature in fahrenheit.
avg_temp_f = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]

avg_temp_c = [((x - 32) * 0.56) for x in avg_temp_f]
```

```python
# x-axis range - months
x_axis = np.arange(1, 13, 1)

x_axis
```

```python

f_handle, = plt.plot(x_axis, avg_temp_f, marker ='+', color='blue', label="Fahrenheit")

c_handle, = plt.plot(x_axis, avg_temp_c, marker='s', color='red', label="Celsius")
```

```python
# Adds a legend and sets its location to the upper right
plt.legend(loc="upper right")
```

```python
# Axis labels
plt.xlabel("Months")
plt.ylabel("Degrees")
```
