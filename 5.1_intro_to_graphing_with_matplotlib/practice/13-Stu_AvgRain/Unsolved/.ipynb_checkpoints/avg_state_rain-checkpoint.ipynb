```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
csv_file = "../Resources/avg_rain_state.csv"

df = pd.read_csv(csv_file)

df.head()
```

```python
plt.bar(df["State"], df["Inches"], color='b', alpha=0.5, align="center")
```

```python
plt.ylabel("Average rainfall (in inches)")
plt.xlabel("State")
plt.title("Average rainfall per state for a given year")
```
