```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
csv_file = "../Resources/avg_rain_state.csv"

df = pd.read_csv(csv_file)

df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(df))
tick_locations = [value+0.4 for value in x_axis]
plt.figure(figsize=(20,3))
plt.bar(x_axis, df["Inches"], color="blue", alpha=0.5, align="edge")
plt.xticks(tick_locations, df["State"], rotation="vertical")
plt.xlabel("State")
plt.ylabel("Amount of Rain (Inches)")
plt.title("Average Rainfall per State")
# Set x and y limits
plt.xlim(-0.25, len(x_axis))
plt.ylim(0, max(df["Inches"])+10)
plt.tight_layout()
```
