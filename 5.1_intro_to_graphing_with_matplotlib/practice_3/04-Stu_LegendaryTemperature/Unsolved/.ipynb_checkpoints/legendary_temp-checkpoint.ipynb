```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Average temperature in fahrenheit
average_temp_f = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]

# Average temperature in celsisus
average_temp_c = [(temp-32) * 0.56 for temp in average_temp_f]

# x axis
x_axis = np.arange(1, 13, 1)
```

```python
# Plot graph
fahrenheit, = plt.plot(x_axis, average_temp_f, marker="+", color="blue", label="Fahrenheit")
celsius, = plt.plot(x_axis, average_temp_c, marker="s", color="red", label="Celsius")
```

```python
# Add legend
plt.legend(loc="best")
```

```python
# Add axes labels and title.
plt.ylabel("Degrees")
plt.xlabel("Months")
plt.title("Monthly Averages for Temperature in New Jersey")
```

```python
# Save figure and show.
plt.savefig("../Images/temp.png")
plt.show()
```
