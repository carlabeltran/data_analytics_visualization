```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Average temperature in fahrenheit.
average_temp_f = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Assign to the x-axis a range of numerical values representing each month of the year.
x_axis = np.arange(1, 13, 1)

x_axis
```

```python
# Plot the degrees Fahrenheit points.
plt.plot(x_axis, average_temp_f)
plt.show()
```

```python
# Use a list comprehension to convert the temperature to degrees Celsius.
average_temp_c = [(temp-32) * 0.56 for temp in average_temp_f]

average_temp_c
```

```python
# Plot the degrees Celsius points.
plt.plot(x_axis, average_temp_c)
plt.show()
```

```python
# Create a third plot with both the degrees Fahrenheit and degrees Celsius points.

plt.plot(x_axis, average_temp_f)
plt.plot(x_axis, average_temp_c)
plt.xlabel("Months")
plt.ylabel("Degrees")
plt.title("Monthly Averages for Temperature in New Jersey")
plt.show()
```
