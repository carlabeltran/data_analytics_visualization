```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create x-axis using numeric value for months. 1 is January, 2 is February, etc.
x_axis = np.arange(1, 13, 1)

x_axis
```

```python
# Average temperature per month in Fahrenheit. This will go on y-axis of graph.
average_temps_f = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]

average_temps_f
```

```python
# Plot average temperature n F.
# Give our graph axis labels
plt.xlabel("Month")
plt.ylabel("Average Temperature (F)")
plt.plot(x_axis, average_temps_f)
```

```python
# Convert averge temps to Celsius.
average_temps_c = [(temp - 32) * 5 / 9 for temp in average_temps_f]

average_temps_c
```

```python
# Plot average temperature n C.
# Give our graph axis labels
plt.xlabel("Month")
plt.ylabel("Average Temperature (C)")
plt.plot(x_axis, average_temps_c)
```

```python
# Plot average temperatures on same graph.
plt.plot(x_axis, average_temps_f)
plt.plot(x_axis, average_temps_c)
```

```python
# Give our graph axis labels
plt.xlabel("Month")
plt.ylabel("Average Temperature")

# Have to plot our chart once again as it doesn't stick after being shown
plt.plot(x_axis, average_temps_f)
plt.plot(x_axis, average_temps_c)
plt.show()
```
