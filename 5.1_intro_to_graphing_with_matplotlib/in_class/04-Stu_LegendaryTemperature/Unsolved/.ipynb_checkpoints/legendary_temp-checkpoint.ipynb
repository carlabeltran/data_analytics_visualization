```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create x-axis using numeric value for months. 1 is January, 2 is February, etc.
x_axis = np.arange(1, 13, 1)

x_axis
```

```python
# Average temperature per month in Fahrenheit. This will go on y-axis of graph.
average_temps_f = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]

average_temps_f
```

```python
# Convert average temps to Celsius.
average_temps_c = [(temp - 32) * 5 / 9 for temp in average_temps_f]

average_temps_c
```

```python
# Assign plots to tuples that stores result of plot

# Each point on the fahrenheit chart is marked by a blue plus
fahrenheit_handle, = plt.plot(x_axis, average_temps_f, marker ='+', color='blue', label="Fahrenheit")

# Each point on the celsius chart is marked by a red square
celsius_handle, = plt.plot(x_axis, average_temps_c, marker='s', color='red', label="Celsius")
```

```python
# Give our graph axis labels
plt.xlabel("Months")
plt.ylabel("Degrees")
```

```python
# Adds a legend and sets its location to the upper right
plt.legend(loc="upper right")
```

```python
# Saves an image of our chart so that we can view it in a folder
plt.savefig("../Images/legendendary_temp.png")
plt.show()
```
