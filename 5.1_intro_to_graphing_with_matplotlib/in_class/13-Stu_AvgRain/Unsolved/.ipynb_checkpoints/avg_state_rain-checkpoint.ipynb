```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
csv_file = "../Resources/avg_rain_state.csv"

df = pd.read_csv(csv_file)

df.head()
```

```python
states = list(df["State"])
states

rainfall_inches = list(df["Inches"])
rainfall_inches

x_axis = np.arange(len(states))
```

```python
plt.figure(figsize=(20,5))
plt.bar(x_axis, rainfall_inches, color='b', alpha=0.5, align="edge")
```

```python
# Create the ticks for our bar chart's x axis
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, states, rotation="vertical")
# Set x and y limits
plt.xlim(-0.25, len(x_axis))
plt.ylim(0, max(df["Inches"])+10)
# Save our graph and show the grap
plt.tight_layout()
plt.savefig("../Images/avg_state_rain.png")
plt.show()
```

```python
plt.title("Average rainfall (in inches) per state")
plt.xlabel("State")
plt.ylabel("Average rainfall (inches)")
```
