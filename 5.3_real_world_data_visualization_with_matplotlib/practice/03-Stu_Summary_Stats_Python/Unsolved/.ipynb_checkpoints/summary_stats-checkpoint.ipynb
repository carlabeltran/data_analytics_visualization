```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
```

```python
# Read in the california housing data set
csv_file = "../Resources/California_Housing.csv"

df = pd.read_csv(csv_file)

df.head()
```

```python
# Determine which measure of central tendency is most appropriate to describe the Population
population = df["Population"]
mean_numpy = np.mean(population)
median_numpy = np.median(population)
mode_scipy = sts.mode(population)

print(f"The mean of the population is {mean_numpy}.")
print(f"The median of the populatio is {median_numpy}.")
print(f"The mode of the population is {mode_scipy}.")
```

```python
# Determine if the house age in California is considered normally distributed
house_age = df["HouseAge"]
plt.hist(house_age)
plt.xlabel('House Age (Years)')
plt.ylabel('Counts')
plt.show()
print(sts.normaltest(house_age.sample(50)))
```

```python
# Determine if there are any potential outliers in the average occupancy in California
ave_occup = df["AveOccup"]

quartiles = ave_occup.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of average occupancy is: {lowerq}")
print(f"The upper quartile of average occupancy is: {upperq}")
print(f"The interquartile range of average occupancy is: {iqr}")
print(f"The the median of average occupancy is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```

```python
# With the potential outliers, what is the lowest and highest median income (in $1000s) observed?
outliers = df.loc[(df["AveOccup"] < lower_bound) | (df["AveOccup"] > upper_bound)]

outliers.head()

lowest_med_income = outliers["MedInc"].min()
highest_med_income = outliers["MedInc"].max()
print(lowest_med_income)
print(highest_med_income)
```

```python
# Bonus - plot the latitude and longitude of the California housing data using Matplotlib,
# color the data points using the median income of the block.
latitude = df["Latitude"]
longitude = df["Longitude"]
med_inc = df["MedInc"]

plt.scatter(latitude, longitude, c=med_inc, alpha=0.3)
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("Latitude and Longitude for California Houses")
clb = plt.colorbar()
clb.set_label("Median Income")
```
