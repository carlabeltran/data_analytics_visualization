```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
```

```python
# Read in the california housing data set
csv_file = "../Resources/California_Housing.csv"

df = pd.read_csv(csv_file)

df.head()
```

```python
# Determine which measure of central tendency is most appropriate to describe the Population
population = df["Population"]

mean = np.mean(population)
print(f"The mean of the population is {round(mean, 2)}")
median = np.median(population)
print(f"The median of the population is {round(median, 2)}")
mode = sts.mode(population)
print(f"The mode of the population is {mode}")
```

```python
# Determine if the house age in California is considered normally distributed
house_age = df["HouseAge"]

plt.hist(house_age)
plt.xlabel("House Age (years)")
plt.ylabel("Counts")
plt.show()

print(sts.normaltest(house_age.sample(50)))
```

```python
# Determine if there are any potential outliers in the average occupancy in California
average_occupancy = df["AveOccup"]
quartiles = average_occupancy.quantile([.25, .5, .75])
lowerq = quartiles[.25]
upperq = quartiles[.75]
median = quartiles[.5]
iqr = upperq = lowerq

print(lowerq)
print(upperq)
print(median)
print(iqr)

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```

```python
# With the potential outliers, what is the lowest and highest median income (in $1000s) observed?
outliers = df.loc[(average_occupancy < lower_bound) | (average_occupancy > upper_bound)]

outliers.head()

lowest_median_income = outliers["MedInc"].min()
highest_median_income = outliers["MedInc"].max()
print(lowest_median_income)
print(highest_median_income)
```

```python
# Bonus - plot the latitude and longitude of the California housing data using Matplotlib, color the data points using the median income of the block.
latitude = df["Latitude"]
longitude = df["Longitude"]
median_income = df["MedInc"]

plt.scatter(latitude, longitude, alpha=0.3, c=median_income)
clb = plt.colorbar()
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("Latitude and Longitude of California Housing")
clb.ax.set_ylabel('Median Income', rotation=270)
```
