```python
# Dependencies
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
```

```python
# Load crime data set into pandas
crime_data = pd.read_csv("../Resources/crime_data.csv")

crime_data.head()
```

```python
# Generate a scatter plot of violent crime rate versus year
x_values = crime_data["year"]
y_values = crime_data["violent crime rate"]
plt.scatter(x_values, y_values)
plt.xlabel("Year")
plt.ylabel("Violent Crime Rate")
plt.show()
```

```python
# Perform a linear regression on violent crime rate versus year
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
```

```python
# Create equation of line to calculate predicted violent crime rate
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
```

```python
# Plot the linear model on top of scatter
year = crime_data.iloc[:,0]
violent_crime_rate = crime_data.iloc[:,3]
plt.scatter(year, violent_crime_rate)
plt.plot(year, regress_values, "--")
plt.annotate(line_eq,(1997, 400), fontsize=15, color="red")
plt.xticks(x_values, rotation=90)
plt.xlabel("Year")
plt.ylabel("Violent Crime Rate")
plt.show()
```

```python
# Repeat plotting scatter and linear model for murder rate versus year
year = crime_data.iloc[:,0]
murder_rate = crime_data.iloc[:,5]
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(year, murder_rate)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(year, murder_rate)
plt.plot(year, regress_values, "--")
plt.annotate(line_eq,(1997, 5), fontsize=15, color="red")
plt.xticks(x_values, rotation=90)
plt.xlabel("Year")
plt.ylabel("Murder Rate")
plt.show()
```

```python
# Repeat plotting scatter and linear model for aggravated assault versus year
year = crime_data.iloc[:,0]
aggravated_assault = crime_data.iloc[:,10]
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(year, aggravated_assault)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(year,aggravated_assault)
plt.plot(year, regress_values, "--")
plt.annotate(line_eq,(1997, 800000), fontsize=15, color="red")
plt.xticks(x_values, rotation=90)
plt.xlabel("Year")
plt.ylabel("Aggravated Assault")
plt.show()
```
