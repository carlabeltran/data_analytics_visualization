```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
csv_file = "Resources/got.csv"
df = pd.read_csv(csv_file)

df.head()
```

```python
# Get attacker and defender data
attacker_series = df["attacker_king"].value_counts()

attacker_df = pd.DataFrame(attacker_series)
print(attacker_df)

defender_series = df["defender_king"].value_counts()

defender_df = pd.DataFrame(defender_series)
print(defender_df)

attacker_defender_combined = attacker_series.to_frame().join(defender_series, how="outer")

attacker_defender_combined

attacker_defender_combined["Number of Battles Participated In"] = attacker_defender_combined.sum(axis=1)

attacker_defender_combined = attacker_defender_combined[["Number of Battles Participated In"]]

attacker_defender_combined
```

```python
# Get total battle data
x_axis = np.arange(len(df))
tick_locations = [value for value in x_axis]
```

```python
# Configure plot and ticks
attacker_defender_combined.plot(kind="bar", color="red")

# Set a title for the chart
plt.title("The Bloodthirst of Kings")
plt.xlabel("King")
plt.ylabel("Number of Battles Participated In")

plt.show()
plt.tight_layout()
```
