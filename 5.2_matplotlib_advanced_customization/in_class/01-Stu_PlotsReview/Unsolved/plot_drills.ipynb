```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATA SET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
plt.bar(gyms, members, facecolor="b", alpha=0.5, align="center")
plt.xlabel("Gym")
plt.ylabel("Number of Members")
plt.title("Number of Members per Gym")
plt.xlim(-0.75, len(gyms)-0.25)
plt.ylim(0, max(members) + 5)
```

```python
# DATA SET 2
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)

print(x_axis)
print(sin)
```

```python
plt.plot(x_axis, sin, color="red")
plt.hlines(0, 0, x_lim, alpha=0.5)
plt.ylabel("sin(x)")
plt.xlabel("Real numbers from 0 to 6")
plt.title("Sin from 0 to 6")
```

```python
# DATA SET 3
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
x_axis = np.arange(0, len(gyms))
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = (0, 0.05, 0, 0)
```

```python
plt.pie(members, labels=gyms, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("Percentage of Memberships by Gym")
plt.axis("equal")
```

```python
# DATA SET 4
x_axis = np.arange(0, 10, 0.1)
times = []
for x in x_axis:
    times.append(x * x + np.random.randint(0, np.ceil(max(x_axis))))
    
print(times)
```

```python
plt.scatter(x_axis, times, marker="o", facecolor="red", edgecolor="black", alpha=0.5)
plt.xlabel("Numbers from 0 to 6")
plt.ylabel("Time (s)")
plt.title("Time it took to do something")
```
