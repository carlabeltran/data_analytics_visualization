```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATA SET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
x_axis = np.arange(0, len(gyms), 1)
tick_locations = [value for value in x_axis]

plt.bar(x_axis, members, color="blue", alpha=0.5)
plt.xticks(tick_locations, gyms)

plt.xlabel("Gyms")
plt.ylabel("Number of Members")
plt.title("Number of Members per Gym")

plt.xlim(-0.75, len(gyms)-0.25)
plt.ylim(0, max(members)+5)
```

```python
# DATA SET 2
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)
```

```python
plt.plot(x_axis, sin, marker="o", linewidth=0, color="red", alpha=0.5)
plt.hlines(0, 0, x_lim, alpha=0.2)

plt.xlabel("Real numbers from 0 to 6")
plt.ylabel("sin(x)")
plt.title("Sin curve")
```

```python
# DATA SET 3
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
x_axis = np.arange(0, len(gyms))
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = (0, 0.05, 0, 0)
```

```python
plt.pie(members, labels=gyms, colors=colors, explode=explode, autopct="%1.1f%%", startangle=140)
plt.title("Percentage of Gym Memberships")
```

```python
# DATA SET 4
x_axis = np.arange(0, 10, 0.1)
times = []
for x in x_axis:
    times.append(x * x + np.random.randint(0, np.ceil(max(x_axis))))
```

```python
plt.scatter(x_axis, times, marker="o", facecolors="red", edgecolors="black", alpha=0.5)
plt.xlabel("Length of Input Array")
plt.ylabel("Time to Sort (s)")
plt.title("Running Time of FakeSort for Sample Input Sizes")
```
